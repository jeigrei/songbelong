```python
from bs4 import BeautifulSoup
```

```python
with open('td_results_0.html', 'r') as f:
    res0 = f.read()
with open('td_results_1.html', 'r') as f:
    res1 = f.read()
```

```python
soup0 = BeautifulSoup(res0, 'lxml')
```

```python
divs = soup0.find_all('div', class_="card__header__title")
```

```python
titles = []
```

```python
for div in divs:
    title = div.text.split('\n')[1].strip()
    titles.append(title)
```

```python
soup1 = BeautifulSoup(res1, 'lxml')
divs = soup1.find_all('div', class_='card__header__title')
```

```python
for div in divs:
    title = div.text.split('\n')[1].strip()
    titles.append(title)
```

```python
titles
```

```python
with open('songs.txt', 'w') as f:
    f.writelines(titles)
```
